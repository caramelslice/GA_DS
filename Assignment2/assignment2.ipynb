```python

# casual users ride more on the weekend
# register users ride more during the week

#Registered riders ride mostly during rush hours
#casual riders like to ride in the afternoon


plt.figure()
groupedwd = bike.groupby(['weekday'])
groupedwd_casual_trendmax = groupedwd.describe().unstack().casual['mean']
groupedwd_casual_trendmax.plot(figsize=(12, 6), label='Casual Riders mean', linewidth=3.0)

groupedwd_reg_trendmax = groupedwd.describe().unstack().registered['mean']
groupedwd_reg_trendmax.plot(figsize=(12, 6), label='Registered Riders mean', linewidth=3.0)

groupedwdtotalmax = groupedwd.describe().unstack().cnt['mean']
groupedwdtotalmax.plot(figsize=(12, 6), label='Total Riders mean', linewidth=3.0) 


plt.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.ylabel('Average Number of riders')
plt.xlabel('Day of the week')
plt.title(r'Ridership maximum by week of the day')
```

```python
plt.figure()
groupedsn = bike.groupby(['season'])
groupedsn_casual_trendmax = groupedsn.describe().unstack().casual['mean']
groupedsn_casual_trendmax.plot(figsize=(12, 6), label='Casual Riders mean', linewidth=3.0)

groupedsn_reg_trendmax = groupedsn.describe().unstack().registered['mean']
groupedsn_reg_trendmax.plot(figsize=(12, 6), label='Registered Riders mean', linewidth=3.0)

groupedsntotalmax = groupedsn.describe().unstack().cnt['mean']
groupedsntotalmax.plot(figsize=(12, 6), label='Total Riders mean', linewidth=3.0) 


plt.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.ylabel('Average Number of riders')
plt.xlabel('Season')
plt.title(r'Ridership by Season')

#always more riders in summer
```

```python
plt.figure()
groupedtmp = bike.groupby(['temp'])
groupedtmp_casual_trendmax = groupedtmp.describe().unstack().casual['mean']
groupedtmp_casual_trendmax.plot(figsize=(12, 6), label='Casual Riders mean', linewidth=3.0)

groupedtmp_reg_trendmax = groupedtmp.describe().unstack().registered['mean']
groupedtmp_reg_trendmax.plot(figsize=(12, 6), label='Registered Riders mean', linewidth=3.0)

groupedtmptotalmax = groupedtmp.describe().unstack().cnt['mean']
groupedtmptotalmax.plot(figsize=(12, 6), label='Total Riders mean', linewidth=3.0) 


plt.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.ylabel('Average Number of riders')
plt.xlabel('Temperature (Normalized temperature in Celsius. The values are divided to 41 (max))')
plt.title(r'Temperature impact on ridership')


#always more riders in summer, especially, people like to ride when it's around 70 farenheit 
```

```python
plt.figure()
groupedwd = bike.groupby(['windspeed'])
groupedwd_casual_trendmax = groupedwd.describe().unstack().casual['mean']
groupedwd_casual_trendmax.plot(figsize=(12, 6), label='Casual Riders mean', linewidth=3.0)

groupedwd_reg_trendmax = groupedwd.describe().unstack().registered['mean']
groupedwd_reg_trendmax.plot(figsize=(12, 6), label='Registered Riders mean', linewidth=3.0)

groupedwd_totalmax = groupedwd.describe().unstack().cnt['mean']
groupedwd_totalmax.plot(figsize=(12, 6), label='Total Riders mean', linewidth=3.0) 


plt.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.ylabel('Average Number of riders')
plt.xlabel('Wind (Normalized wind speed. The values are divided to 67 (max))')
plt.title(r'Wind Speed impact on ridership')

#people are unlikely to ride when it's more windy
```

```python
plt.figure()
groupedsn = bike.groupby(['weathersit']).aggregate(sum)
groupedsn['registered'].plot(figsize=(12, 6), label='Registered Riders', linewidth=3.0) 
groupedsn['casual'].plot(figsize=(12, 6), label='Casual Riders', linewidth=3.0) 
plt.ylabel('Number of riders')
plt.title(r'Weather impact on ridership')
plt.legend(loc='upper center', shadow=True, fontsize='x-large')
#people are more likely to ride when the weather is Clear, Few clouds, Partly cloudy, Partly cloudy
# It's very unlikely anybody would ride in Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
```
